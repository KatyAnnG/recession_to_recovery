```python
def extract_month_from_filename(df):
    months = ['January', 'February', 'March', 'April', 'May', 'June',
              'July', 'August', 'September', 'October', 'November', 'December']

    for month_name in months:
        mask = df['file_name'].str.contains(month_name, case=False)
        df.loc[mask, 'month'] = month_name[:3].lower()
```

```python
cols = ['year', 
        'month', 
        'Region', 
        'Total Adults']

df2 = df2[cols]
```

```python
df2['total adults'] = df2['total adults'].str.replace(',', '')
```

```python
df2['total adults'].astype('int64')
```

```python
df2.columns = df2.columns.str.lower()
```

```python
df_grouped = df2.groupby(['year', 'region'])[['total adults']].median()
```

```python
df_grouped.reset_index(inplace=True)
```

```python
df_grouped
```

```python
df_totals = df_grouped.groupby(['year'])[['total adults']].sum()
```
