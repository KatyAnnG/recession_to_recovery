#### Population percentage change

```python
df6 = pd.read_csv('./population_percentage_change_2002-2006.csv')
```

```python
df6 = df6[df6['Census Year'] > 1996]
df6
```

```python
df7 = pd.read_csv('./population_percentage_change_2011-2022.csv')
df7
```

```python
df7.rename(columns={'CensusYear':'Census Year', 'County and City':'County'}, inplace=True)
```

```python
def process_pop_data(df, merge_columns):
    
    df.columns = df.columns.str.lower()
    df.rename(columns={'census year':'year','county':'region'}, inplace=True)
    
    label1 = 'Population'
    label2 = 'Actual change since previous census'
    label3 = 'Percentage change since previous census'
    
    df1 = df[df['statistic label'] == label1]
    df2 = df[df['statistic label'] == label2]
    df3 = df[df['statistic label'] == label3]
    
    df1.rename(columns={'value': f'{label1}' }, inplace=True)
    df2.rename(columns={'value': f'{label2}' }, inplace=True)
    df3.rename(columns={'value': f'{label3}' }, inplace=True)
    
    df1 = df1.drop(columns=['unit', 'statistic label'])
    df2 = df2.drop(columns=['unit', 'statistic label'])
    df3 = df3.drop(columns=['unit', 'statistic label'])
    
    df_temp = df1.merge(df2, how='left', on=merge_columns)
    df_full = df_temp.merge(df3, how='left', on=merge_columns)
    
    df_full.columns = df_full.columns.str.lower()
    
    return df_full

merge_columns_df6 = ['year', 'region']
merge_columns_df7 = ['year', 'region', 'sex']

merged_df6 = process_pop_data(df6, merge_columns_df6)
merged_df7 = process_pop_data(df7, merge_columns_df7)
```

```python
merged_df6
```

```python
merged_df7 = merged_df7[merged_df7['sex'] == 'Both sexes']
merged_df7
```

```python
merged_df7.pop('sex')
```

```python
merged_df7
```

```python
percentage_change_full = merged_df6.merge(merged_df7, how='outer', on=['year',
                                                                'region',
                                                                'population',
                                                                'actual change since previous census',
                                                                'percentage change since previous census'])
```

```python
percentage_change_full#.to_csv('./pop_percentage_change_full.csv')
```
